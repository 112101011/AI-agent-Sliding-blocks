# Generate Maze


A maze can be visualized as an arrangement of cells in a rectangular $m \times m$ grid with walls between some pairs of cells. 

(a)  Write a program that uses randomized depth-first search to generate a maze. The randomized depth-first search procedure is as follows: Starting from a given cell (say (0,0)), this algorithm produces a path that visits each cell in the grid according to the following recursive procedure.
- If the current cell $C$ has a neighbouring cell that is not yet visited, chose one of such cells (say $C'$) at random and remove the wall between these two cells. Repeat with $C'$ as the current cell.
- If all neighbouring cells of the current cell $C$ are already visited, then backtrack to the last cell $\widehat{C}$ with a neighbouring cell $\widehat{C}'$ that is not yet visited and repeat with $\widehat{C}'$ as the current cell.

A sample output of a $3\times 3$ maze is along with the graph adjacency representation is as shown below. The bottom left corner of the maze is $(0,0)$ and top right corner of the maze is $(2,2)$.

```python
import random
class Node:
    '''This class is for a node a singly linked list'''
    def __init__(self, vertex):
        self.vertex = vertex
        self.next = None
    
class Maze_problem:
    '''This class is for maze problem'''
    def __init__(self, m):
        self.m = m
        Adjlists = dict()
        for j in range(m):
            for i in range(m):
                Adjlists[(i, j)] = [None, None] # stack operation on linked list maintaining both head and tail
                
        self.Adjlists = Adjlists
    
    # insertion in the beginning of Linked list as per the example
    def __insert(self, source, cell): 
        head, tail = self.Adjlists[source]
        temp = Node(cell)
        if head is None:
            self.Adjlists[source] = [temp, temp]
        else:
            temp.next = head
            self.Adjlists[source][0] = temp
        return
    
    # DFS randomized
    def __randomized_DFS(self, source, visited):
        neighbours = []
        count = 0
        i, j = source[0], source[1]
        
        if i != self.m - 1 and (visited[self.m - 1 - (i + 1)][j] is False):
            neighbours.append((i + 1, j))
            count += 1
            
        if i != 0 and (visited[self.m - 1 - (i - 1)][j] is False):
            neighbours.append((i - 1, j))
            count += 1
        
        if j != 0 and (visited[self.m - 1 - i][j - 1] is False):
            neighbours.append((i, j - 1))
            count +=1
        
        if j != self.m - 1 and (visited[self.m - 1 - i][j + 1] is False):
            neighbours.append((i, j + 1))
            count += 1
        
        while count != 0:
            
            # Randomly selecting the unvisted neighbour
            cell = neighbours.pop(random.randrange(count))
            count -= 1
            p, q = cell[0], cell[1]
            
            # Marking the cell as visited
            visited[self.m - 1 - p][q] = True
            
            # Adding edge(undirected) in adjacency lists:
            self.__insert(source, cell)
            self.__insert(cell, source)
            
            # Recursion
            self.__randomized_DFS(cell, visited)
            
            # removing the vertices which are visited in recursion
            temp = []
            count = 0
            for cell in neighbours:
                p, q = cell[0], cell[1]
                if visited[self.m - 1 - p][q] is False:
                    temp.append(cell)
                    count += 1
            neighbours = temp
    
    # Function for generating maze
    def generate_maze(self):
        visited = [[False for i in range(self.m)]for j in range(self.m)]
        source = (0, 0)
        i, j = source[0], source[1]
        visited[self.m - 1 - i][j] = True
        self.__randomized_DFS(source, visited)
    
    def print_list(self):
        for cell in self.Adjlists:
            print(cell, ":", end = " ")
            head = self.Adjlists[cell][0]
            while head is not None:
                print(head.vertex, end = "")
                head = head.next
                if head is not None:
                    print(", ", end = "")
            print()
        return
    
    def __print_maze(self, Horz, Ver):
        flag = True
        i = 0
        j = 0
        for z in range(2 * self.m + 1):
            if flag is True:
                for ele in Horz[i]:
                    print("+{0}".format(ele), end = "")
                i += 1
                flag = False
                print('+', end = "")
            else:
                for ele in Ver[j]:
                    print(ele, " ", end = " ")
                j += 1
                flag = True
            print()
    
    def __update(self, source, visited, Horz, Ver):
        # print(source)
        # self.__print_maze(Horz, Ver)
        
        i, j = source[0], source[1]
        visited[self.m - 1 - i][j] = True
        
        head = self.Adjlists[source][0]
        while head is not None:
            cell = head.vertex
            p, q = cell[0], cell[1]
            if visited[self.m - 1 - p][q] is False:
                # there is a path from source to cell
                if p != i:
                    if p > i:
                        Horz[(self.m - 1 - i)][j] = '   '
                    else:
                        Horz[(self.m - 1 - i) + 1][j] = '   '
                else:
                    if q > j:
                        Ver[self.m - 1 - i][j + 1] = ' '
                    else:
                        Ver[self.m - 1 - i][j] = ' '
                self.__update(cell, visited, Horz, Ver)
            head = head.next
            
    def getMaze(self):
        Horz = [['---' for j in range(self.m)] for i in range(self.m + 1)]
        Ver = [['|' for j in range(self.m + 1)] for i in range(self.m)]
        
        visited = [[False for i in range(self.m)] for j in range(self.m)]
        source = (0, 0)
        self.__update(source, visited, Horz, Ver)
        
        print("FINAL MAZE:")
        self.__print_maze(Horz, Ver)

m = int(input())
obj = Maze_problem(m)
obj.Adjlists
obj.generate_maze()
obj.print_list()
print()
obj.getMaze()
```

(d)  Write a program to do A$^*$ search on a $m \times m$ maze given in adjacency representation to find a route from the source cell $(0,0)$ to the destination cell $(m-1,m-1)$. Use the Manhattan heuristic for A$^*$ search. The Manhattan distance between two cells of the maze $(i,j)$ and $(k,\ell)$ where $i,j,k,\ell \in \{0,1\ldots, m-1\}$ is $|i-k| + | j-\ell|$. Output the route and the number of cells explored.

```python
'''
The folowing A star algo works for any source and goal
Heuristic = Manhattan distance
Euclidian distance can be the other heuristic
Manhattan distance would be the actual shortest path possible if there are no obstacles

Tie breaking policy:
if f(n) is same for both cells then choose the one with less h(n), if h(n) is also same we can chose any cell

'''

from queue import PriorityQueue

# MOD function
def MOD(value):
    if value >= 0:
        return value
    return -1 * value

# h(n)
def Manhattan_heuristic(cell, goal):
    i, j = cell[0], cell[1]
    k, l = goal[0], goal[1]
    return MOD(i - k) + MOD(j - l)

def A_star(obj, source, goal):
    parent = dict()
    q = PriorityQueue()          # Min heap

    g_n = 0 # for source
    h_n = Manhattan_heuristic(source, goal)
    parent[source] = (None, g_n)
    q.put((g_n + h_n, h_n, source))
    explored = 0
    
    while q.empty() is not True:
        cell = q.get()[2]
        parent_g_n = parent[cell][1]
        
        head = obj.Adjlists[cell][0]
        if cell == goal: # late goal test, (here late or early doesn't matter since the edges are unweighted)????
            isPath = True
            break
        g_n = parent_g_n + 1
        while head is not None:
            adjVer = head.vertex
            head = head.next
            if head is None:
                explored += 1
            if parent.get(adjVer, False) is False or g_n < parent[adjVer][1]:
                parent[adjVer] = (cell, g_n)
                h_n = Manhattan_heuristic(adjVer, goal)
                q.put((g_n + h_n, h_n, adjVer))
    path = ''
    if isPath is True:
        node = goal
        path = path + f"{goal}"
        while node != source: # time: O(d)
            node = parent[node][0]
            path = f"{node}-->" + path
    return path, explored, isPath

# source node
source = (0, 0)

# goal node
goal = (obj.m - 1, obj.m - 1)
path, explored, isPath = A_star(obj, source, goal)

if isPath is True:
    print(f"There is a path from {source} to {goal}")
    print("PATH:", path)
else:
    print(f"There is no path from {source} to {goal}")
print('Number of cells explored =', explored)
```

# Sliding Blocks

Consider the sliding blocks puzzle where we are given a $3 \times 3$ grid of blocks with each block containing a unique integer between 0 and 8. An example configuration is given below.

1 2 3

7 8 5

0 6 4

At each step, the block containing 0 can swap places with an adjacent block. 


(a)  Use A$^*$ search to start from any given initial configuration and reach the  goal configuration 

 1 2 3

 4 5 6

 7 8 0

with the sum of Manhattan distances of the blocks from their goal positions as the heuristic. The Manhattan distance of a pair of blocks occupying the integer $n$ at the locations $(i,p)$ and $(j,q)$ (where $i,j,p,q \in \{0,\ldots, 2\}$ and $n \in \{0,\ldots,8 \}$) is given by $ |i-j| + |p-q|$. Print the total number of steps taken to reach the goal and the blocks configuration at each step. 

```python
from queue import PriorityQueue

class sliding_blocks:
    '''This class is sliding blocks puzzle'''
    def __init__(self, m, goal):
        self.m = m
        goal_map = dict()
        for i in range(self.m):
            for j in range(self.m):
                goal_map[goal[i][j]] = (self.m - 1 - i, j)
        self.goal_map = goal_map
    
    def MOD(self, value):
        if value >= 0:
            return value
        return -1 * value
    
    # Heuristic 1
    def Manhattan_distance(self, present_state):
        h_n = 0
        for i in range(self.m):
            for j in range(self.m):
                p_row = self.m - 1 - i  # present state row
                p_col = j               # present state column
                
                key = present_state[i][j]
                g_row, g_col = self.goal_map[key] # goal row, goal column
                h_n += MOD(p_row - g_row) + MOD(p_col - g_col)
        return h_n
    
    # Heuristic 2
    def misplaced_blocks(self, present_state):
        h_n = 0
        for i in range(self.m):
            for j in range(self.m):
                p_row = self.m - 1 - i  # present state row
                p_col = j               # present state column
                
                key = present_state[i][j]
                g_row, g_col = self.goal_map[key]
                if p_row != g_row or p_col != g_col:
                    h_n += 1
        return h_n
    
    # Heuristic 3
    def zero_Manhattan_distance(self, present_state):
        h_n = 0
        for i in range(self.m):
            for j in range(self.m):
                if present_state[i][j] == 0:
                    p_row = self.m - 1 - i  # present state row
                    p_col = j               # present state column
                    
                    key = present_state[i][j]
                    g_row, g_col = self.goal_map[key] # goal row, goal column
                    h_n += MOD(p_row - g_row) + MOD(p_col - g_col)
                    return h_n
        return h_n
    
    # Goal test 
    def goal_test(self, present_state):
        for i in range(self.m):
            for j in range(self.m):
                key = present_state[i][j]
                p_row = self.m -1 - i     # present state row
                p_col = j                 # present state col
                
                g_row, g_col = self.goal_map[key]
                
                if p_row != g_row or p_col != g_col:
                    return False
        return True
    
    def find_neighbours(self, present_state):
        neighbours = []
        zero_row = -1
        zero_col = -1
        
        flag = False
        for i in range(self.m):
            for j in range(self.m):
                row = self.m - 1 - i
                col = j
                
                if present_state[i][j] == 0:
                    zero_row = row
                    zero_col = col
                    flag = True
            if flag is True:
                break
        
        # North Direction
        if zero_row != self.m - 1:
            next_state = [list(row) for row in present_state]
            next_state[self.m -1 - zero_row][zero_col], next_state[self.m - 1 - (zero_row + 1)][zero_col] = next_state[self.m - 1 - (zero_row + 1)][zero_col], next_state[self.m -1 - zero_row][zero_col] 
            next_state = tuple(tuple(row) for row in next_state)
            neighbours.append(next_state)
        
        # South Direction
        if zero_row != 0:
            next_state = [list(row) for row in present_state]
            next_state[self.m -1 - zero_row][zero_col], next_state[self.m - 1 - (zero_row - 1)][zero_col] = next_state[self.m - 1 - (zero_row - 1)][zero_col], next_state[self.m -1 - zero_row][zero_col] 
            next_state = tuple(tuple(row) for row in next_state)
            neighbours.append(next_state)
        
        # West Direction
        if zero_col != 0:
            next_state = [list(row) for row in present_state]
            next_state[self.m -1 - zero_row][zero_col], next_state[self.m - 1 - zero_row][zero_col - 1] = next_state[self.m - 1 - zero_row][zero_col - 1], next_state[self.m -1 - zero_row][zero_col] 
            next_state = tuple(tuple(row) for row in next_state)
            neighbours.append(next_state)
        
        # East Direction
        if zero_col != self.m - 1:
            next_state = [list(row) for row in present_state]
            next_state[self.m -1 - zero_row][zero_col], next_state[self.m - 1 - zero_row][zero_col + 1] = next_state[self.m - 1 - zero_row][zero_col + 1], next_state[self.m -1 - zero_row][zero_col] 
            next_state = tuple(tuple(row) for row in next_state)
            neighbours.append(next_state)
        
        return neighbours
    
    # Astar algo
    def a_star(self, initial_state, ques):
        q = PriorityQueue()
        parent = dict()
        
        # explored = 0
        h_n = self.heuristic(ques, initial_state)
        g_n = 0
        f_n = g_n + h_n
        
        parent[initial_state] = [g_n, None]
        q.put((f_n, initial_state))
        action_cost = 1
        
        isPath = False
        while q.empty() is not True:
            f_n, present_state = q.get()
            parent_g_n = parent[present_state][0]
            
            if self.goal_test(present_state) is True:
                isPath = True
                break
            neighbours = self.find_neighbours(present_state)
            g_n = action_cost + parent_g_n
            for next_state in neighbours:
                g_n = action_cost + parent_g_n
                if parent.get(next_state, False) is False or g_n < parent[next_state][0] :
                    parent[next_state] = [g_n, present_state]
                    h_n = self.heuristic(ques, present_state)
                    f_n = g_n + h_n
                    q.put((f_n, next_state))
        
        path = ''
        if isPath is True:
            node = goal
            temp = ''
            for row in node:
                temp = temp + f'{row}\n'    
            path = temp + '\n'
            while node != source:
                node = parent[node][1]
                temp = ''
                i = 0
                for row in node:
                    if i == 1:
                        temp = temp + f'{row} -->\n'
                    else:
                        temp = temp + f'{row}\n'
                    i += 1
                temp += '\n'
                path = temp + path
        return path, len(parent), isPath
    
    def heuristic(self, ques, present_state):    # Works similar to API
        if ques == 1:
            return self.Manhattan_distance(present_state)
        elif ques == 2:
            return self.misplaced_blocks(present_state)
        elif ques == 3:
            return self.zero_Manhattan_distance(present_state)
```
